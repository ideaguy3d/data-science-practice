# imports

```python
import pandas as pd
```

```python
# training data
df = pd.read_csv("data/Training_Data-Household_survey_features.csv")

# keep every other row (rows with even index)
df_halved = df.iloc[::2].reset_index(drop=True)

# write out the smaller CSV
df_halved.to_csv("training_features_household_survey_halved.csv", index=False)
```

```python
# test data
df = pd.read_csv("data/Test_Data-Household_survey_features.csv")

# keep every other row (rows with even index)
df_halved = df.iloc[::2].reset_index(drop=True)

# write out the smaller CSV
df_halved.to_csv("Test_features_household_survey_halved.csv", index=False)
```
